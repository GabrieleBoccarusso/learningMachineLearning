```python
# all together:
# 0. import matplotlib and get it ready for plotting in jupyter notebook
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare the data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. setup plot
fig, ax = plt.subplots(figsize=(10,10)) # creating figure and axis
                            # width and height of the figure
    
# 3. plot data
ax.plot(x, y)

# 4. customize plot
ax.set(title = "simple plot",
       xlabel = "x_axis",
       ylabel = "y_axis")

```

```python
# settingup the example data
x = [5, 10, 15, 20]
y = [50, 100, 150, 200]

# setting up the figure
figure, axis = plt.subplots()

# customizing data
axis.set(title = "example plot", xlabel = "x_axis", ylabel = "y_axis")

# displaying the data
axis.plot(x, y);

# saving the data
figure.savefig("example.png")
```

```python
# make a plot from a dictionary
nut_butter_prices = { "almond butter": 1000,
                      "Peanut butter": 800,
                      "Cashew butter": 1200}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title= "Dan's Nut Butter Store", ylabel="price ($)");
```

```python
car_prices = { "Audi": 22000,
               "Honda": 8000,
               "Ariel": 45000}

fig, ax = plt.subplots()
ax.set(title= "car seller prices", ylabel="price ($)")
ax.bar(car_prices.keys(), car_prices.values());
```

```python
# horizontal plot
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
# barh (bar horizontal) accept only lists as parameters
```

```python
# making in histogram
# randm create given random numbers
# in the normal distribution
values = np.random.randn(100)
figure, axis = plt.subplots()
axis.hist(values);
```

```python
#making a figure with various plots in
fig , ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                              ncols = 2,
                                              figsize = (10, 5))
# every single plot
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
#making a figure with various plots in
fig , ax = plt.subplots(nrows = 2,
                                              ncols = 2,
                                              figsize = (10, 5))
# every single plot
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

## plotting from panda dataframes

```python
import pandas as pd
# make a data frame
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
# removing any special character (Price is currently a string)
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '')
car_sales
```

```python
# removing last two zeros
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
# adding a column filled with dates
car_sales["sales date"] = pd.date_range("2/1/2021", periods = len(car_sales))
car_sales
```

```python
car_sales["total sales"] = car_sales["Price"].astype(int).cumsum()
car_sales.tail()
```

```python
car_sales.plot(x="sales date", y="total sales");
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter");
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
bb_players = pd.read_csv("bruh.csv")
bb_players[:10]
```

```python
print(bb_players.keys())
```

```python
bb_players.rename(columns = {' "Team"': 'team',
                             ' "Position"': 'Position',
                             ' "Height(inches)"': 'Height (inches)',
                             ' "Weight(lbs)"': 'Weight (lbs)',
                             ' "Age"': 'Age'}, 
                  inplace = True)
bb_players[:10]
```

```python
bb_players['Weight (lbs)'].dtype
```

```python
# a['nums']=a['nums'][a['nums']!=''].astype(float)
bb_players['Weight (lbs)'] = bb_players['Weight (lbs)'][bb_players['Weight (lbs)']!=' ""'].astype(float)
bb_players.head()
```

```python
bb_players.head(5)
```

```python
bb_players['Weight (lbs)'].dtype
```

```python
# over_50 = heart_disease[heart_disease["age"] > 50]
catchers = bb_players[bb_players['Position'] == ' "Catcher"']
catchers.tail(10)
```

```python
len(catchers)
```

```python
bb_players["Age"].hist();
```

```python
# creating a figure with more than 1 subplots
figure, (first_plot, second_plot) = plt.subplots(nrows = 2,
                            ncols = 1,
                            figsize = (12, 20))

scatter_plot = first_plot.scatter(x = bb_players['Age'],
                                  y = bb_players['Weight (lbs)'],
                                  c = bb_players['Height (inches)'])
first_plot.set(title="correlations between baseball players weight and age",
               xlabel="age of players",
               ylabel="weight of players")
first_plot.legend(*scatter_plot.legend_elements(), title = 'Height in inches');

# putting a line on the average data
first_plot.axhline(y = bb_players['Weight (lbs)'].mean(), 
                   color = 'r', 
                   linestyle = '-')

# second plot
second_plot.hist(x = bb_players['Age'], bins = 20)
second_plot.set(title = "Numbers of players with a certain age",
                xlabel="age of players",
                ylabel="number of players");
```

## customizing matplotlib plots

```python
plt.style.available
```

```python
# changing plot style
plt.style.use('seaborn-deep')
# creating a figure with more than 1 subplots
figure, (first_plot, second_plot) = plt.subplots(nrows = 2,
                            ncols = 1,
                            figsize = (12, 20))

scatter_plot = first_plot.scatter(x = bb_players['Age'],
                   y = bb_players['Weight (lbs)'],
                   c = bb_players['Height (inches)'],
                   cmap = 'winter')
first_plot.set(title="correlations between baseball players weight and age",
               xlabel="age of players",
               ylabel="weight of players")
first_plot.legend(*scatter_plot.legend_elements(), title = 'Height in inches');

# putting a line on the average data
first_plot.axhline(y = bb_players['Weight (lbs)'].mean(), 
                   linestyle = '-')

# second plot
second_plot.hist(x = bb_players['Age'], bins = 20)
second_plot.set(title = "Numbers of players with a certain age",
                xlabel="age of players",
                ylabel="number of players");
```
